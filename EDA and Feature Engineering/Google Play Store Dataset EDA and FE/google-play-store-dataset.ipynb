#### **2. Size**

```python
# check for missing values
df_copy['Size'].isnull().sum()
```

#### <span style="color:red;">Obs: </span> There are no missing values present in the 'Size' column.

```python
df_copy['Size'].unique()
```

- 19M = 19000000
- 19k = 19000
- 19M = 19000k

```python
# replace millions with thousands
df_copy['Size'] = df_copy['Size'].str.replace('M', '000')

# replace thousands with blank
df_copy['Size'] = df_copy['Size'].str.replace('k', '')

# replace Varies with device with nan
df_copy['Size'] = df_copy['Size'].replace('Varies with device', np.nan)
```

#### Converting the datatype of the 'Size' column to float

```python
df_copy['Size'] = df_copy['Size'].astype(float)
```

```python
df_copy.info()
```

#### <span style="color:red;">Obs: </span> From the above info we can see that we have successfully converted the datatype of the 'Size' column from object to float.

```python
# check again for missing values
df_copy['Size'].isnull().sum()
```

#### <span style="color:red;">Obs: </span> There are 1695 missing values present in the 'Size' column. We will fix those in the Feature Engineering part further.

#### **3. Installs and Price**

```python
df.head()
```

```python
df_copy['Installs'].unique()
```

```python
df_copy['Price'].unique()
```

```python
# replace all the special characters '+',',','$' to blank
chars_to_remove = ['+',',','$']            # item
cols_to_clean = ['Installs','Price']       # cols
 
for item in chars_to_remove:
    for cols in cols_to_clean:
        df_copy[cols] = df_copy[cols].str.replace(item, '')
```

```python
df_copy['Installs'].unique()
```

```python
df_copy['Price'].unique()
```

#### <span style="color:red;">Obs: </span> From the above info we can see that all the special characters have been successfully replaced with blanks.

#### Converting the datatype of 'Installs' column to int and 'Price' columns to float

```python
df_copy['Installs'] = df_copy['Installs'].astype(int)
df_copy['Price'] = df_copy['Price'].astype(float)
```

```python
df_copy.info()
```

#### <span style="color:red;">Obs: </span> From the above info we can see that the dataype conversion is successful.

#### **4. Last Updated**

```python
df_copy.head(2)
```

```python
df_copy['Last Updated'].unique()
```

```python
# convert the dataype of the 'Last Updated' column to datetime
df_copy['Last Updated'] = pd.to_datetime(df_copy['Last Updated'])

# get the day from 'Last Updated' column
df_copy['Day'] = df_copy['Last Updated'].dt.day

# get the month from 'Last Updated' column
df_copy['Month'] = df_copy['Last Updated'].dt.month

# get the year from 'Last Updated' column
df_copy['Year'] = df_copy['Last Updated'].dt.year
```

#### Q2. Which are top 10 App categories

#### Q3. Which Category has largest number of installations?

```python
# Group and process data
df2 = (df_copy.groupby('Category')
       ['Installs'].sum()
       .div(1e9)  # Convert to billions
       .sort_values(ascending = False)  # sort in desc
       .reset_index()
       .head(10))

# Plot
plt.figure(figsize=(14, 10))
sns.barplot(x = 'Installs', y = 'Category', data = df2, palette = 'viridis')
plt.xlabel('No. of Installations in Billions', size = 20)
plt.ylabel('')
plt.title("Most Popular Categories in Play Store", size = 20)
plt.show()
```

#### <span style="color:red;">Insights: </span>
1. Out of all the categories "GAME" has the most number of Installations.
2. With almost 35 Billion Installations, GAME is the most popular Category in Google App Store.

#### Q4. What are the Top 5 most installed Apps in Each popular Categories?

```python
# Group and aggregate data
dfa = df_copy.groupby(['Category', 'App'])['Installs'].sum().reset_index().sort_values('Installs', ascending=False)

# Categories to visualize
apps = ['GAME', 'COMMUNICATION', 'PRODUCTIVITY', 'SOCIAL']

# Plotting
plt.figure(figsize=(40, 30))
sns.set_context("poster")
sns.set_style("darkgrid")

for i, app in enumerate(apps):
    df3 = dfa[dfa['Category'] == app].head(5)
    plt.subplot(4, 2, i + 1)
    sns.barplot(x='Installs', y='App', data=df3, palette='magma')
    plt.xlabel('Installation in Millions')
    plt.title(app, size=20)

plt.tight_layout()
plt.subplots_adjust(hspace=0.3)
plt.show()
```

#### <span style="color:red;">Insights: </span>
- Most popular game is Subway Surfers.
- Most popular communication app is Skype.
- Most popular productivity app is Google Drive.
- Most popular social app is Facebook.

#### Q5. How many apps are there on Google Play Store which get 5 ratings?

```python
rating = df_copy.groupby(['Category','Installs','App'])['Rating'].sum().sort_values(ascending = False).reset_index()

top_rating_apps = rating[rating.Rating == 5.0]
print('Number of 5 rated apps:', top_rating_apps.shape[0])

print("Top most rated app:")
top_rating_apps.head(1)
```

#### <span style="color:red;">Insights: </span>
- There are 271 five rated apps on Google Play Store.
- Top most rated app is 'CS & IT Interview Questions' from 'Family' Category.
